```
import networkx as nx
import matplotlib.pyplot as plt
import statistics as st
import numpy as np
import random as random

"""def get_vertices(G):
  return G.number_of_edges()

def get_diameter(G):
  return nx.diameter(G)
"""


def get_degree(G):
  dict0=dict(nx.degree(G))
  result=dict0.items()
  data = list(result)
  arr0 = np.array(data)
  arr1=[row[1] for row in arr0]
  print("degree Standard Deviation")
  print(np.std(arr1))
  print("degree Mean:")
  print(np.mean(arr1))
  print("degree Mean 2m/n:")
  n=G.number_of_nodes()
  m=G.number_of_edges()
  print((2*m)/n)

def get_elements(G):
  print("Number of Edges:")
  print(G.number_of_edges())
  print("Diameter:")
  print(nx.diameter(G))
  print("Radius:")
  print(nx.radius(G))



```

# Barab√°si-Albert model

```
def generate_ba(n, m0, m):
    """ Function to generate graph networks using
    Watts-Strogatz Model


    Returns
    -------
    random WS graph
    """
    return nx.barabasi_albert_graph(n,m0, m)

def plot_ba(G, n):
    """ Plots the Watts-Strogatz graph

    Parameters
    ----------
    G: graph
    n: int
      Number of nodes

    Returns
    -------
    None
    """
    # Put the nodes in a circular shape
    pos = nx.circular_layout(G)
    pos = nx.spring_layout(G, iterations=400, seed=42)
    nx.draw(G, pos, node_color=range(400), node_size=50, cmap=plt.cm.Blues)
    plt.show()

#Here BA-Scale free network with N nodes were generated
```

```
m=3
m0=3
n=400

G_ba = generate_ba(n, m0, m)
plot_ba(G_ba, n)
get_elements(G_ba)
get_degree(G_ba)
```

# Ego model Based on Max Degree on BA

```
degree_list=[]
max=0
max_node=0
for i in range (50):
  degree_list.append(nx.degree(G_ba)[i])
  if degree_list[i] > max:
    max=degree_list[i]
    max_node=i


hub_ego = nx.ego_graph(G_ba, max_node)

# Draw graph
pos = nx.spring_layout(hub_ego)  # Seed layout for reproducibility
nx.draw(hub_ego, pos, node_color="b", node_size=50, with_labels=False)

# Draw ego as large and red
options = {"node_size": 300, "node_color": "r"}
nx.draw_networkx_nodes(hub_ego, pos, nodelist=[max_node], **options)
plt.show()
```
